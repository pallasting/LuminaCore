```
import torch
import lumina.nn as lnn
import os

# Enable Rust Backend
os.environ["LUMINA_USE_RUST"] = "1"

# Create a photonic layer
layer = lnn.OpticalLinear(784, 128, hardware_profile="lumina_nano_v1")

# Forward pass
x = torch.randn(32, 784)
output = layer(x)

print(f"Output shape: {output.shape}")
print(f"Using Rust Backend: {os.environ.get('LUMINA_USE_RUST')}")
```

## Benchmark Rust vs PyTorch (Simulation)

```
import time

# Warm up
_ = layer(x)

start = time.time()
for _ in range(100):
    _ = layer(x)
end = time.time()

print(f"Time for 100 forward passes: {end - start:.4f} seconds")
```
